```python
 # Week 1: Introduction to Machine Learning & Python Libraries
 import numpy as np
 import pandas as pd
 print("Week 1: NumPy and Pandas Basics")
 a = np.array([1, 2, 3, 4])
 print("NumPy Array:", a)
 print("Mean:", np.mean(a))
 print("Standard Deviation:", np.std(a))
 data = {'Name': ['Anji', 'Balu', 'Kalyan'], 'Marks': [85, 90, 95]}
 df = pd.DataFrame(data)
 print("Pandas DataFrame:\n", df)
```

```python
 # Week 2: Linear Regression - California Housing Dataset
 from sklearn.datasets import fetch_california_housing
 from sklearn.linear_model import LinearRegression
 from sklearn.model_selection import train_test_split
 print("\nWeek 2: Linear Regression")
 housing = fetch_california_housing()
 X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target,
 test_size=0.2)
 model = LinearRegression().fit(X_train, y_train)
 print("Model Score:", model.score(X_test, y_test))
```

```python
 # Week 3: Logistic Regression - Pima Indians Diabetes Dataset
 import seaborn as sns
 from sklearn.linear_model import LogisticRegression
 from sklearn.metrics import accuracy_score
 print("\nWeek 3: Logistic Regression")
 df = sns.load_dataset("iris")
 df = df[df['species'].isin(['setosa', 'versicolor'])]
 X = df.drop(columns='species')
 y = df['species'].apply(lambda x: 1 if x == 'versicolor' else 0)
 X_train, X_test, y_train, y_test = train_test_split(X, y)
 model = LogisticRegression().fit(X_train, y_train)
 pred = model.predict(X_test)
 print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# Week 4: K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier
print("\nWeek 4: K-Nearest Neighbors")
iris = sns.load_dataset("iris")
X = iris.drop("species", axis=1)
y = iris["species"]
X_train, X_test, y_train, y_test = train_test_split(X, y)
knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
print("KNN Accuracy:", knn.score(X_test, y_test))
```

```python
 # Week 5: Decision Trees
 from sklearn.tree import DecisionTreeClassifier, plot_tree
 import matplotlib.pyplot as plt
print("\nWeek 5: Decision Trees")
 tree = DecisionTreeClassifier().fit(X_train, y_train)
 plt.figure(figsize=(10, 6))
 plot_tree(tree, filled=True, feature_names=X.columns, class_names=tree.classes_)
 plt.title("Decision Tree")
 plt.show()
```

```python
# Week 6: Support Vector Machine (SVM)
from sklearn.svm import SVC
print("\nWeek 6: Support Vector Machine")
svm = SVC(kernel='linear').fit(X_train, y_train)
print("SVM Accuracy:", svm.score(X_test, y_test))
```

```python
 # Week 7: Naive Bayes Classifier
 from sklearn.naive_bayes import MultinomialNB
 from sklearn.feature_extraction.text import CountVectorizer
 print("\nWeek 7: Naive Bayes - Spam Detection")
 texts = ["Congratulations, you've won!", "Call now!", "Are we still meeting?", "Let'scatch up tomorrow"]
 labels = [1, 1, 0, 0]
 vectorizer = CountVectorizer()
 X = vectorizer.fit_transform(texts)
 model = MultinomialNB().fit(X, labels)
 test = vectorizer.transform(["Win a free ticket", "See you later"])
 print("Predictions:", model.predict(test))
```

```python
# Week 8: K-Means Clustering
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
print("\nWeek 8: K-Means Clustering")
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```

```python
 # Week 9: Principal Component Analysis (PCA)
 from sklearn.decomposition import PCA
 print("\nWeek 9: Principal Component Analysis")
 pca = PCA(n_components=2).fit(X)
 X_pca = pca.transform(X)
 plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
 plt.title("PCA Projection")
 plt.show()
```

```python
 # Week 10: Random Forest Classifier - Titanic Dataset
 from sklearn.ensemble import RandomForestClassifier
 import seaborn as sns
 print("\nWeek 10: Random Forest - Titanic Dataset")
 titanic = sns.load_dataset('titanic').dropna()
 X = titanic[['age', 'fare', 'pclass']]
 y = titanic['survived']
 X_train, X_test, y_train, y_test = train_test_split(X, y)
 rf = RandomForestClassifier().fit(X_train, y_train)
 print("Random Forest Accuracy:", rf.score(X_test, y_test))
```

```python
# Week 11: Model Evaluation Metrics
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
print("\nWeek 11: Model Evaluation")
y_pred = rf.predict(X_test)
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
 # Week 12: Output Visualization
 import seaborn as sns
 print("\nWeek 12: Visualization")
 sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
 plt.title("Confusion Matrix Heatmap")
 plt.xlabel("Predicted")
 plt.ylabel("Actual")
 plt.show()
```
